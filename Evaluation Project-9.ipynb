```python
x= df.drop('TotalCharges', axis=1)
y= df['TotalCharges']
```

```python
x.shape
```

```python
y.shape
```

# Scaling Using Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler

mn=MinMaxScaler()
x=mn.fit_transform(x)
```

# Importing Model

```python
import sklearn
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import Ridge, Lasso
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
ne = KNeighborsRegressor(n_neighbors=2)
lr= LinearRegression()
dt = DecisionTreeRegressor()
rf= RandomForestRegressor(n_estimators=100)
ab= AdaBoostRegressor()
rd= Ridge()
mod = [ne, lr, dt, rf, ab, rd]
mod
```

```python
for j in mod:
    for i in range(0,100):
        xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.22, random_state=i)
        j.fit(xtrain,ytrain)
        pred_train = j.predict(xtrain)
        pred_test= j.predict(xtest)

        print(f"For Model {j} At random state {i}, the training accuracy is: {r2_score(ytrain,pred_train)}")
        print(f"For Model {j} At random state {i}, the testing accuracy is: {r2_score(ytest,pred_test)}")
        print('\n')
```

As per above visualization we found model Decision Tree Regressor is best at random state 5

```python
print("Mean Squared error", mean_squared_error(ytest,pred_test))
print("Mean Absolute Error", mean_absolute_error(ytest,pred_test))
```

```python
pred_train = dt.predict(xtrain)
pred_test= dt.predict(xtest)
print(r2_score(ytest,pred_test))
```

# Grid Search CV

```python
from sklearn.model_selection import GridSearchCV

model = DecisionTreeRegressor()
param_grid = {'criterion' :['mse','mae'], "max_depth":[10,20,50,100], 'max_features':['auto','sqrt','log2'],'random_state':[10,20,22,42,45]}
xg_grid = GridSearchCV(estimator = model, param_grid = param_grid)
print(xg_grid)
xg_grid.fit(xtrain, ytrain)
```

```python
print(xg_grid.best_score_)
print(xg_grid.best_estimator_)
print(xg_grid.best_params_)
```

As per Grid Search CV we found the best accuracy when criterion=mse, random_state= 10,max_features:auto,max_depth:10

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.22, random_state=10)
DTR = DecisionTreeRegressor(criterion='mse',max_depth=10, max_features='auto',random_state=10)
DTR.fit(xtrain,ytrain)
pred_train =DTR.predict(xtrain)
pred_test= DTR.predict(xtest)
print(r2_score(ytest,pred_test))
print("Mean Squared Error", mean_squared_error(ytest,pred_test))
print("Mean Absolute Error", mean_absolute_error(ytest,pred_test))
```

```python
DTR.score(xtrain,ytrain)
```

```python
DTR.score(xtrain,pred_train)
```

# Cross Validation of Model

```python
from sklearn.model_selection import cross_val_score
for i in range (2,10):
    score= cross_val_score(DTR,x,y,cv=i)
    print("At cv value at", i,"the cv score is",score)
    print(score.mean())
    print(score.std())
```
